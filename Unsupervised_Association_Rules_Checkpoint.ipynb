```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
print(df.shape)
```

```python
toy_dataset = [
    ['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
    ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']
]

# Encode transactions
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
df_checkpoint = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
df_checkpoint.head()
```

```python
transactions = []
for i in range(df_checkpoint.shape[0]):
    transactions.append([str(df_checkpoint.values[i,j]) for j in range(df_checkpoint.shape[1]) if str(df_checkpoint.values[i,j]) != 'nan'])
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)
```

```python

frequent_itemsets = apriori(df_trans, min_support=0.02, use_colnames=True)


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)


rules.sort_values('lift', ascending=False).head(10)
```

Items frequently bought together (high confidence and lift) indicate natural product pairings.

Example: {Bread} â†’ {Butter}, {Milk} â†’ {Cereal}.


Stock high-lift item pairs together to reduce stockouts and improve customer convenience.

Promoting popular combinations can increase the average basket size and overall revenue.
